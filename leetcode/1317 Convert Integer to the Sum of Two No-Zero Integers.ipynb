```python
class Solution:
    def getNoZeroIntegers(self, n: int):
        l, r = 0, n
        print(f'{l:5}, {r:5}')
        while True:
            if str(l).count("0")==0 and str(r).count("0")==0:
                print(f'{l:5}, {r:5}')
                return l, r
            l += 1
            r -= 1
            print(f'{l:5}, {r:5}')
```

```python
n = 10101
Solution().getNoZeroIntegers(n)
```
