```python
class Solution:
    def largestTriangleArea(self, points):
        from itertools import combinations
        return max(self.area(*three_points) for three_points in combinations(points, 3))
    def area(self, p, q, r):
        """
        compute area of triangle generated by three vectors p, q, r
        """
        return abs(p[0]*q[1]+q[0]*r[1]+r[0]*p[1] - (p[1]*q[0]+q[1]*r[0]+r[1]*p[0])) / 2
```
